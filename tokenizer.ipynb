```python
import re
from tokenizer import SimpleTokenizerV1, SimpleTokenizerV2
```

```python
with open("the-verdict.txt", "r") as f:
    raw_text = f.read()
print("Total no. of characters:",len(raw_text))
```

```python
#Split the data into individual words, free from whitespace and punctuation
words = [item for item in re.split(r'([,.:;?_!"()\']|--|\s)', raw_text) if item.strip()]
print("Total no. of words:",len(words))
words
```

```python
#Creating token ids
all_words = sorted(set(words))
size_of_vocab = len(all_words)
print("Size of vocabulary:",size_of_vocab)
vocab = {word: token for token, word in enumerate(all_words)}
inverse_vocab = {token: word for token, word in enumerate(all_words)}
```

```python
#Use the created tokenizer, check its workings
tokenizer = SimpleTokenizerV1(vocab_to_id=vocab)
tokenizerv2 = SimpleTokenizerV2(vocab_to_id=vocab)

encoded_text = tokenizer.encode(raw_text)
decoded_text = tokenizer.decode(encoded_text)
decoded_text
```
